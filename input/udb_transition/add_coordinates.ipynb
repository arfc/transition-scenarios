```python
# connect sqlite db
import sqlite3 as lite
import numpy as np
```

```python
con = lite.connect('coordinates.sqlite')
con.row_factory = lite.Row
cur = con.cursor()
```

```python
# read file
newfile = []
nextup = False
with open('./reactors.xml', 'r') as f:
    lines = f.readlines()
    for line in lines:
        if '<name>' in line:
            start = line.index('>')
            end = line.index('</')
            name = line[start + 1 : end]
            if name[-1].isdigit():
                name = name[:-2]
            x = cur.execute('SELECT * FROM reactors_coordinates '
                            'WHERE name LIKE "%%%s%%" '
                            'AND country = "United States of America"' %name).fetchall()
            if name == 'YANKEE NPS':
                name = 'YANKEE ROWE'
            if name == 'HADDAM NECK':
                name = 'Connecticut'
            if name == 'DAVIS BESSE':
                name = 'DAVIS'

            if len(x) >= 2:
                print(name)
            for row in x:
                long = row['Long']
                lat = row['Lat']
        if nextup:
            newfile.append('\t<longitude>%f</longitude>\n' %long)
            newfile.append('\t<latitude>%f</latitude>\n' %lat)
            nextup = False
        if '<power_cap>' in line:
            nextup = True
        newfile.append(line)
            
        
```

```python
for line in newfile:
    print(line)
```

```python
with open('reactor_with_coordinates.xml', 'w') as wf:
    for line in newfile:
        wf.write(line)
```
