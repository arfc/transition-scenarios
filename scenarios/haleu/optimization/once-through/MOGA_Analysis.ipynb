# MOGA Results
This notebook contains the results from each of the Multi-objective Genetic Algorithm (MOGA) 
problems run as part of Amanda Bachmann's dissertation work. The problems aim to minimize waste 
and SWU to produce HALEU in the transition from the US LWR fleet to the X-energy Xe-100, NuScale VOYGR,
and USNC MMR.

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from pygmo import hypervolume
import csv
```

## Once-through scenario

```python
once_through_pareto = pd.read_csv(f"moga_pareto.dat", sep="\s+", header = None)
once_through_pareto = once_through_pareto.rename(columns={0:'mmr_share', 1:'xe100_share',
                                                         2:'voygr_share', 3:'lwr',
                                                         4:'mmr_burnup', 5:'xe100_burnup',
                                                         6:'haleu_swu', 7:'waste',
                                                         8:'total_share'})
```

```python
reference = [6e9, 2e8]
hv = hypervolume(once_through_pareto[['haleu_swu','waste']].values)
hv.compute(reference)
```

```python
plt.plot(once_through_pareto['haleu_swu'], once_through_pareto['waste'], linestyle='None', marker = 'o')
plt.xlabel('HALEU SWU (kg-SWU)')
plt.ylabel('Waste (kg)')
plt.savefig('once_through_pareto.pdf')
```

```python
plt.plot(once_through_pareto['xe100_share'], once_through_pareto['voygr_share'], linestyle='None', marker = 'o')
plt.xlabel("Xe-100 build share (%)")
plt.ylabel("VOYGR build share (%)")
```
