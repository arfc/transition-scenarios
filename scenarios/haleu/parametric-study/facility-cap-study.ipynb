### LEU Enrichment Facility Caps Effects

```python
# No cap on Fabrication Facility Cap
caps = [70000000, 60000000, 50000000, 10000000, 5000000, 2000000,
        1500000, 1000000, 500000, 100000, 10000, 1000, 100, 10, 1, 0]

fig = plt.figure()
ax1 = fig.add_subplot(111)
for cap in caps:
    file = 'united_states_xe100_' + str(cap) + '_None.sqlite'
    energy = tm.get_electricity(file)
    ax1.plot(energy['Year'], energy['Energy'], label= str(cap))
    
#plot reactors
xe_reactors = tm.rx_commission_decommission(file, xe_nonlwr)
xe_reactors = tm.add_year(xe_reactors)
xe_reactors['xe_total'] = xe_reactors['Xe-100'].cumsum()
ax2 = ax1.twinx()
ax2.plot(xe_reactors['Year'], xe_reactors['xe_total'], 'r', label='Xe-100 Reactors')
#Rate of reactor deployment is independent of the facility caps
for tl in ax2.get_yticklabels():
    tl.set_color('r')
#ax2.set_ylabel('Energy (MWe-yr)')

ax1.set_xlabel('Year')
ax1.set_ylabel('Energy (MWe-y)')
ax1.legend(loc=(1.1,0))
plt.title('Effect of Enrichment Cap, no fabrication cap')
```
